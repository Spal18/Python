```python
#5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap.
#Ans.

import seaborn as sns
import numpy as np
#Import the pandas module
import pandas as pd

# Generate synthetic dataset with correlated features
np.random.seed(0)
n_samples = 100
x1 = np.random.rand(n_samples)
x2 = 0.7 * x1 + np.random.randn(n_samples) * 0.2
x3 = -0.5 * x1 + np.random.randn(n_samples) * 0.3
data = pd.DataFrame({'x1': x1, 'x2': x2, 'x3': x3})

# Calculate the correlation matrix
correlation_matrix = data.corr()

# Create a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
#3. Using the sales data, generate a heatmap to visualize the variation in sales across different months and days.
'''np.random.seed(20)
data = {'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
        'Day': np.random.choice(range(1, 31), 100),
        'Sales': np.random.randint(1000, 5000, 100)}
df = pd.DataFrame(data)'''

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

np.random.seed(20)
data = {
 'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
 'Day': np.random.choice(range(1, 31), 100),
 'Sales': np.random.randint(1000, 5000, 100)
 }
df = pd.DataFrame(data)

# Create a pivot table to aggregate sales by month and day
pivot_table = df.pivot_table(values='Sales', index='Month', columns='Day', aggfunc='sum')

# Generate the heatmap
plt.figure(figsize=(12, 6))
sns.heatmap(pivot_table, annot=True, fmt=".0f", cmap='viridis')
plt.title('Sales Heatmap by Month and Day')
plt.xlabel('Day')
plt.ylabel('Month')
plt.show()
```

```python
#4. Using the given x and y data, generate a 3D surface plot to visualize the function.
'''x = np.linspace(-5, 5, 100)
  y = np.linspace(-5, 5, 100)
  x, y = np.meshgrid(x, y)
  z = np.sin(np.sqrt(x**2 + y**2))
  data = {'X': x.flatten(),'Y': y.flatten(),'Z': z.flatten() }
  df = pd.DataFrame(data)'''


import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

data = {
 'X': x.flatten(),
 'Y': y.flatten(),
 'Z': z.flatten()
}
df = pd.DataFrame(data)

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

ax.plot_trisurf(df['X'], df['Y'], df['Z'], cmap='viridis')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D Surface Plot of z = sin(sqrt(x^2 + y^2))')

plt.show()
```

```python
#1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.?

from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np


# Create data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create a figure
p = figure(title="Sine Wave", x_axis_label="x", y_axis_label="sin(x)")

# Add a line glyph
p.line(x, y, legend_label="Sine Wave", line_width=2)

# Show the plot
show(p)
```

```python
# 3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.
#fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
#counts = [20, 25, 30, 35]

from bokeh.io import output_notebook
from bokeh.plotting import figure, show

fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

# Replace 'plot_height' with 'height' to set the figure height
p = figure(x_range=fruits, height=350, title="Fruit Counts",
           toolbar_location=None, tools="")

p.vbar(x=fruits, top=counts, width=0.9)

p.xgrid.grid_line_color = None
p.y_range.start = 0

show(p)
```

```python
#4. Create a Bokeh histogram to visualize the distribution of the given data.
# data_hist = np.random.randn(1000)
# hist, edges = np.histogram(data_hist, bins=30)


from bokeh.plotting import figure, show
import numpy as np

data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)

p = figure(title="Histogram", x_axis_label="Value", y_axis_label="Frequency")
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:],
       fill_color="navy", line_color="white", alpha=0.5)

show(p)
```
